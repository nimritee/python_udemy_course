```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# NORMAL DISTRIBUTION

```python
from scipy.stats import norm
```

```python
d = np.random.randn(10000)
```

```python
data = np.arange(1,1000,100)
pdf = norm.pdf(data , loc = 50 , scale = 40 )
```

```python
d = np.random.normal(loc=3.0,scale=50, size=(10000))
fig, ax = plt.subplots(figsize = (10,5))
ax.hist(d, bins=100, color='y')
ax.set_title("Random generated data")
ax.set_xlabel('Bins')
ax.set_ylabel('Count')
ax.plot()
```

```python
data = np.arange(1,10,0.1)
pdf = norm.pdf(data , loc = 5.3 , scale = 1 )
 
#Visualizing the distribution
 
sb.set_style('whitegrid')
sb.histplot(pdf, color = 'b')
plt.xlabel('Heights')
plt.ylabel('Probability Density')
```

# EXPONENTIAL DISTRIBUTION

```python
# Using exponential() method
ed = np.random.exponential(5, 1000)
  
count, bins, ignored = plt.hist(ed, 7, density = True, color='darkred')
plt.show()
```

```python
ed1 = np.random.exponential(104.5, 10000)
  
count, bins, ignored = plt.hist(ed1, 8, density = True, color='orange')
plt.show()
```

# UNIFORM DISTRIBUTION

```python
#Importing required libraries
from scipy.stats import uniform

#taking random variables from Uniform distribution
data = uniform.rvs(size = 1000, loc = 0, scale=8) #uniform distribution from range 0-8
fig, ax = plt.subplots(figsize = (10,5))
ax = sb.histplot(data, bins = 50, color = 'c')
ax.set(xlabel = 'interval')
plt.show()
```

```python
data = uniform.rvs(size = 100000, loc = 0, scale=5) #uniform distribution from range 0-8
fig, ax = plt.subplots(figsize = (10,5))
ax = sb.histplot(data, bins = 2000, color = 'r')
plt.ylim(10, 100)
ax.set(xlabel = 'interval')
plt.show()
```

# BINOMIAL DISTRIBUTION

```python
from scipy.stats import binom
```

```python
pb = binom(n = 40, p = 0.6)
 
x = np.arange(0,20)
pmf = pb.pmf(x)
 
#Visualizing the distribution
plt.vlines(x ,0, pb.pmf(x), colors='y', linestyles='-', lw=5)
plt.ylabel('Probability')
plt.xlabel('Intervals')
plt.show()
```

```python
pb = binom(n = 20, p = 0.7)
 
x = np.arange(10,50)
pmf = pb.pmf(x)
 
#Visualizing the distribution
plt.vlines(x ,0, pb.pmf(x), colors='g', linestyles='-', lw=5)
plt.ylabel('Probability')
plt.xlabel('Intervals')
plt.show()
```

# BERNOULI DISTRIBUTION

```python
from scipy.stats import bernoulli
 
#Applying the bernoulli class
data = bernoulli.rvs(size = 1000 , p = 0.8)
 
#Visualizing the results
sb.displot(data, shrink=2 , color = 'pink')
plt.show()
```

```python
data = bernoulli.rvs(size = 100 , p = 0.6)
 
#Visualizing the results
sb.displot(data, shrink=0.8 , color = 'magenta')
plt.show()
```

# POISSON DISTRIBUTION

```python
from scipy.stats import poisson
 
#Applying the poisson class methods
x = np.arange(0,100)
pmf = poisson.pmf(x,20)
 
#Visualizing the results
plt.vlines(x ,0, pmf, colors='violet', linestyles='-', lw=6)
plt.ylabel('Probability')
plt.xlabel('intervals')
plt.show()
```

```python
#Applying the poisson class methods
x = np.arange(0,10)
pmf = poisson.pmf(x,5)
 
#Visualizing the results
plt.vlines(x ,0, pmf, colors='maroon', linestyles='-', lw=6)
plt.ylabel('Probability')
plt.xlabel('intervals')
plt.show()
```
